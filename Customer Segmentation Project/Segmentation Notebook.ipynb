```python
#2
#Standarizing Numeric Variables (Age + Income)
sc = StandardScaler()
df1[['Income','Age']] = sc.fit_transform(df1[['Income', 'Age']])
df1
X = df1
```

```python
#3

#Determining Number of Clusters (Elbow Method)
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
inertias = []

for i in range(1, 11):
    km = KMeans(n_clusters = i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(X)
    inertias.append(km.inertia_)
    
print("\nInertia Plot: Within Cluster SSE of Different Cluster Amounts")
plt.plot(range(1,11), inertias, marker='o', color = 'skyblue')
plt.xlabel('Number of Clusters')
plt.ylabel('Cluster inertia (within-cluster SSE)')
plt.xticks(range(1,11))
plt.tight_layout()
plt.show()
```

```python
#Colours for Silhouette Plots
distinct_colors = ['skyblue', 'powderblue', 'lightsteelblue', 'cadetblue', 'aliceblue']
```

```python

#Silhouette Plot (k = 4)
from sklearn.metrics import silhouette_samples
from matplotlib import cm
km = KMeans(n_clusters = 4,
                init='k-means++',
                n_init=10,
                max_iter=300,
            tol=1e-04,
                random_state=0)
y_km = km.fit_predict(X)
print(y_km)

cluster_labels = np.unique(y_km)
print(cluster_labels)

n_clusters =cluster_labels.shape[0]

silhouette_vals = silhouette_samples(X, y_km, metric = 'euclidean')

distinct_colors = ['skyblue', 'powderblue', 'lightsteelblue', 'cadetblue', 'aliceblue']


#plotting silhouette values

y_ax_lower, y_ax_upper = 0, 0
yticks = []


for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    
    color = distinct_colors[i % len(distinct_colors)]
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    
    yticks.append((y_ax_lower + y_ax_upper)/2.)
    y_ax_lower += len(c_silhouette_vals)

silhouette_avg = np.mean(silhouette_vals)
print("\nSilhouette Plot: 4 Clusters")

plt.axvline(silhouette_avg, color="darkblue", linestyle="--")

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
plt.show()
```
